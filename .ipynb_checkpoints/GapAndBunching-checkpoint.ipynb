```python
import pandas as pd
import numpy as np
```

```python
# load
bus_df = pd.read_excel("data/raw/ttc_bus/ttc-bus-delay-data-2024.xlsx")

test_df = bus_df.dropna(subset=["Min Delay", "Min Gap", "Route"]).copy()
test_df["Min Delay"] = pd.to_numeric(test_df["Min Delay"], errors="coerce")
test_df["Min Gap"] = pd.to_numeric(test_df["Min Gap"], errors="coerce")
test_df = test_df.dropna(subset=["Min Delay", "Min Gap"])
```

```python
# 1. Does Min Gap vary a lot within the same route?
gap_stats = test_df.groupby("Route")["Min Gap"].agg(["count", "mean", "std"])
gap_stats = gap_stats[gap_stats["count"] >= 20]
gap_stats["coef_of_variation"] = gap_stats["std"] / gap_stats["mean"]
print("Median coefficient of variation across routes:", round(gap_stats["coef_of_variation"].median(), 2))
 
# 2. How strongly is Min Gap correlated with Min Delay?
overall_corr = test_df["Min Delay"].corr(test_df["Min Gap"])
print("Correlation between Min Delay and Min Gap:", round(overall_corr, 3))
 
# 3. What is the typical Min Gap / Min Delay ratio?
ratio_df = test_df[test_df["Min Delay"] > 0].copy()
ratio_df["ratio"] = ratio_df["Min Gap"] / ratio_df["Min Delay"]
print("Median Min Gap / Min Delay ratio:", round(ratio_df["ratio"].median(), 2))

# 4. Best-fit relationship: Min Gap = slope * Min Delay + intercept
slope, intercept = np.polyfit(test_df["Min Delay"], test_df["Min Gap"], 1)
print(f"Best fit: Min Gap = {slope:.2f} * Min Delay + {intercept:.2f}")
```
