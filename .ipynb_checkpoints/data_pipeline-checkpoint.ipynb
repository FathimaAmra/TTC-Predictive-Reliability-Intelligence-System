```python
from pathlib import Path
import pandas as pd
import numpy as np 
from sklearn.neighbors import BallTree 
```

```python
# CONFIG
DATA_DIR = Path("data")

TTC_DIR = DATA_DIR / "ttc_bus"
WEATHER_DIR = DATA_DIR / "weather"
GTFS_DIR = DATA_DIR / "ttc_gtfs"
ROAD_DIR = DATA_DIR / "road_restrictions"
```

```python
# LOAD DATA

# TTC Bus Delays
bus_frames = []

for file in TTC_DIR.iterdir():
    if file.suffix == ".csv":
        bus_frames.append(pd.read_csv(file))

    elif file.suffix == ".xlsx":
        bus_frames.append(pd.read_excel(file))

bus_df = pd.concat(bus_frames, ignore_index=True)

# Weather
weather_df = pd.read_csv(WEATHER_DIR / "toronto_weather_2022_2025.csv")

# Road Restrictions
road_df = pd.concat([pd.read_csv(file, skiprows=1) for file in ROAD_DIR.glob("*.csv")], ignore_index=True)

# GTFS
routes_df = pd.read_csv(GTFS_DIR / "routes.txt")
trips_df = pd.read_csv(GTFS_DIR / "trips.txt")
stops_df = pd.read_csv(GTFS_DIR / "stops.txt")
stop_times_df = pd.read_csv(GTFS_DIR / "stop_times.txt")
shapes_df = pd.read_csv(GTFS_DIR / "shapes.txt")
```

```python
print("\nBus")
print(bus_df.shape)
print(bus_df.columns.tolist())

print("\nWeather")
print(weather_df.shape)
print(weather_df.columns.tolist())

print("\nRoad")
print(road_df.shape)
print(road_df.columns.tolist())

print("\nRoutes")
print(routes_df.shape)
print(routes_df.columns.tolist())

print("\nTrips")
print(trips_df.shape)
print(trips_df.columns.tolist())

print("\nStops")
print(stops_df.shape)
print(stops_df.columns.tolist())

print("\nStop Times")
print(stop_times_df.shape)
print(stop_times_df.columns.tolist())

print("\nshapes")
print(shapes_df.shape)
print(shapes_df.columns.tolist())
```

#### TTC Bus Delay

```python
#add cleaning code 

bus_df["Route"] = bus_df["Route"].astype(str)

bus_df["Min Delay"] = pd.to_numeric(bus_df["Min Delay"], errors="coerce")
bus_df["Date"] = pd.to_datetime(bus_df['Date'], errors="coerce")
bus_df["day_of_week"] = bus_df["Date"].dt.dayofweek

bus_df["hour"] = pd.to_datetime(
    bus_df["Time"],
    format="%H:%M",
    errors="coerce"
).dt.hour

bus_df["month"] = bus_df["Date"].dt.month
bus_df["year"] = bus_df["Date"].dt.year

bus_df["is_weekend"] = (
    bus_df["day_of_week"] >= 5
).astype(int)

bus_df = bus_df.dropna(subset=["Min Delay", "Route"])

bus_df = bus_df[bus_df["Route"].str.lower() != "nan"]

route_stats = (
    bus_df.groupby("Route")
    .agg(
        total_delay=("Min Delay", "sum"),
        avg_delay=("Min Delay", "mean"),
        incident_count=("Min Delay", "count")
    )
    .reset_index()
)

priority_routes = [
    "7", "24", "25", "29", "35",
    "39", "60", "100", "165",
    "924", "925", "929"
]

remaining_routes = route_stats[
    ~route_stats["Route"].isin(priority_routes)
]

top_remaining = remaining_routes.sort_values(
    by="total_delay",
    ascending=False
).head(38)

top_routes = top_remaining["Route"].tolist()

final_routes = priority_routes + top_routes

bus_df = bus_df[
    bus_df["Route"].isin(final_routes)
].copy()

#display
final_routes_df = route_stats[
    route_stats["Route"].isin(final_routes)
].copy()

final_routes_df["Route"] = pd.Categorical(
    final_routes_df["Route"],
    categories=final_routes,
    ordered=True
)

final_routes_df = final_routes_df.sort_values("Route")

final_routes_df = final_routes_df.sort_values(
    by="total_delay",
    ascending=False
)

print(final_routes_df)
```

#### GTFS

```python
# Routes Cleaning
routes_df = routes_df[[
    "route_id",
    "route_short_name",
    "route_long_name"
]].copy()

routes_df["route_short_name"] = (
    routes_df["route_short_name"]
    .astype(str)
    .str.strip()
)

routes_df["route_id"] = (
    routes_df["route_id"]
    .astype(str)
)

routes_df = routes_df.drop_duplicates()

# ---------------------------
# TRIPS
# ---------------------------
trips_df = trips_df[[
    "trip_id",
    "route_id",
    "direction_id",
    "shape_id"
]].copy()

trips_df["trip_id"] = trips_df["trip_id"].astype(str)
trips_df["route_id"] = trips_df["route_id"].astype(str)

trips_df = trips_df.drop_duplicates()

# ---------------------------
# STOPS
# ---------------------------
stops_df = stops_df[[
    "stop_id",
    "stop_name",
    "stop_lat",
    "stop_lon"
]].copy()

stops_df["stop_lat"] = pd.to_numeric(
    stops_df["stop_lat"],
    errors="coerce"
)

stops_df["stop_lon"] = pd.to_numeric(
    stops_df["stop_lon"],
    errors="coerce"
)

stops_df = stops_df.dropna()

# ---------------------------
# STOP TIMES
# ---------------------------
stop_times_df = stop_times_df[[
    "trip_id",
    "stop_id",
    "arrival_time",
    "departure_time",
    "stop_sequence"
]].copy()

stop_times_df["trip_id"] = (
    stop_times_df["trip_id"]
    .astype(str)
)

stop_times_df["stop_id"] = (
    stop_times_df["stop_id"]
    .astype(str)
)

stop_times_df = stop_times_df.drop_duplicates()

# ---------------------------
# SHAPES (optional)
# ---------------------------
shapes_df["shape_pt_lat"] = pd.to_numeric(
    shapes_df["shape_pt_lat"],
    errors="coerce"
)

shapes_df["shape_pt_lon"] = pd.to_numeric(
    shapes_df["shape_pt_lon"],
    errors="coerce"
)

shapes_df = shapes_df.dropna()


# =========================================================
# CHECK RESULTS
# =========================================================

print("Routes:", routes_df.shape)
print("Trips:", trips_df.shape)
print("Stops:", stops_df.shape)
print("Stop Times:", stop_times_df.shape)
print("Shapes:", shapes_df.shape)
```

```python
#Feature 1: Number of stops per route
route_stops = (
    stop_times_df[["trip_id", "stop_id"]]
    .merge(
        trips_df[["trip_id", "route_id"]],
        on="trip_id"
    )
)

stop_count = (
    route_stops.groupby("route_id")
    ["stop_id"]
    .nunique()
    .reset_index(name="stop_count")
)


#Feature 2: Route length (km)
route_shapes = trips_df[
    ["route_id", "shape_id"]
].drop_duplicates()

shape_length = (
    shapes_df.groupby("shape_id")
    ["shape_dist_traveled"]
    .max()
    .reset_index()
)

shape_length.rename(
    columns={
        "shape_dist_traveled":
        "route_length_km"
    },
    inplace=True
)

route_length = route_shapes.merge(
    shape_length,
    on="shape_id",
    how="left"
)

route_length = (
    route_length.groupby("route_id")
    ["route_length_km"]
    .max()
    .reset_index()
)
```

```python
# Clean stop coordinates
# CLEAN GTFS STOP DATA 
stops_small = stops_df[['stop_id', 'stop_name', 'stop_lat', 'stop_lon']].dropna() 
stops_small['stop_lat'] = pd.to_numeric(stops_small['stop_lat'], errors='coerce') 
stops_small['stop_lon'] = pd.to_numeric(stops_small['stop_lon'], errors='coerce') 
stops_small = stops_small.dropna() 


routes_df = routes_df[["route_id", "route_short_name", "route_long_name"]]

bus_df["Route"] = bus_df["Route"].astype(str)

routes_df["route_short_name"] = (
    routes_df["route_short_name"]
    .astype(str)
)

bus_df = bus_df.merge(
    routes_df[
        [
            "route_id",
            "route_short_name",
            "route_long_name"
        ]
    ],
    left_on="Route",
    right_on="route_short_name",
    how="left"
)

selected_route_ids = routes_df[
    routes_df["route_short_name"].isin(final_routes)
]["route_id"].astype(str)

trips_df = trips_df[
    trips_df["route_id"].isin(selected_route_ids)
].copy()

print(
    bus_df[
        ["Route", "route_long_name"]
    ].head(10)
)
```

#### Road Restrictions

```python

# ========================================================= 
# 2. CLEAN ROAD RESTRICTIONS (FILTER IMPORTANT ROADS) 
# ========================================================= 
# =========================================================
# FILTER ACTIVE ROAD RESTRICTIONS
# =========================================================

road_df['start_dt'] = pd.to_datetime(
    road_df['StartTime'],
    unit='ms',
    errors='coerce'
)

road_df['end_dt'] = pd.to_datetime(
    road_df['EndTime'],
    unit='ms',
    errors='coerce'
)

today = pd.Timestamp.now()
future = today + pd.Timedelta(days=7)

road_small = road_df[
    (road_df['RoadClass'].isin([
        'Major Arterial Road',
        'Minor Arterial Road',
        'Collector'
    ]))
    &
    (road_df['start_dt'] <= future)
    &
    (road_df['end_dt'] >= today)
][[
    'ID',
    'Road',
    'Latitude',
    'Longitude',
    'start_dt',
    'end_dt'
]].dropna()

road_small['Latitude'] = pd.to_numeric(
    road_small['Latitude'],
    errors='coerce'
)

road_small['Longitude'] = pd.to_numeric(
    road_small['Longitude'],
    errors='coerce'
)

road_small['Latitude'] = pd.to_numeric(road_small['Latitude'], errors='coerce') 
road_small['Longitude'] = pd.to_numeric(road_small['Longitude'], errors='coerce') 
road_small = road_small.dropna() 
# ========================================================= 
# 3. BALLTREE (SPATIAL MATCHING) 
# ========================================================= 

stops_small = stops_df[
    (~stops_df['stop_name'].str.contains(
        'Station|Platform',
        case=False,
        na=False
    ))
][['stop_id', 'stop_name', 'stop_lat', 'stop_lon']].dropna()

stop_coords = np.radians(stops_small[['stop_lat', 'stop_lon']].values) 
road_coords = np.radians(road_small[['Latitude', 'Longitude']].values) 
tree = BallTree(road_coords, metric='haversine') 
radius = 150 / 6371000 # 150 meters 
indices, distances = tree.query_radius( stop_coords, r=radius, return_distance=True ) 

# ========================================================= 
# 4. BUILD RAW STOP ↔ ROAD MATCHES 
# ========================================================= 
results = []

for i, stop in enumerate(stops_small.itertuples()):

    seen = set()

    for j, idx in enumerate(indices[i]):

        road_id = road_small.iloc[idx]['ID']
        road_name = road_small.iloc[idx]['Road']

        if road_id in seen:
            continue

        seen.add(road_id)

        dist_m = distances[i][j] * 6371000

        results.append({
            "stop_id": stop.stop_id,
            "stop_name": stop.stop_name,
            "restriction_id": road_id,
            "road": road_name,
            "distance_m": round(dist_m, 1)
        })
raw_df = pd.DataFrame(results).drop_duplicates() 

# ========================================================= 
# 5. REMOVE DUPLICATE STOP POINTS (IMPORTANT FIX) 
# ========================================================= 
# keep closest match per stop-restriction pair 
raw_df = raw_df.sort_values('distance_m').drop_duplicates( subset=['stop_id', 'restriction_id'] ) 

# ========================================================= 
# 6. GTFS: STOP → ROUTE MAPPING 
# ========================================================= 
stop_routes = ( stop_times_df[['stop_id', 'trip_id']] 
    .merge(trips_df[['trip_id', 'route_id']], on='trip_id', how='left') [['stop_id', 'route_id']] .drop_duplicates() ) 

# ========================================================= 
# 7. MERGE ROUTE INFO INTO ROAD IMPACT DATA 
# ========================================================= 
df = raw_df.merge(stop_routes, on='stop_id', how='left') 
df = df.dropna(subset=['route_id'])


# =========================================================
# 8. REMOVE DUPLICATE STOP-ROUTE-RESTRICTION RECORDS
# =========================================================
df = df.drop_duplicates(
    subset=['stop_id', 'route_id', 'restriction_id']
)

# =========================================================
# 9. BUILD FEATURES
# =========================================================
# Create project_id from restriction_id
def get_project_id(x):
    parts = str(x).split('-')
    if len(parts) > 1:
        return '-'.join(parts[:-1])
    return str(x)

# Add project_id to raw_df
raw_df['project_id'] = raw_df['restriction_id'].apply(get_project_id)

# Add project_id to df
df = df.merge(
    raw_df[['restriction_id', 'project_id']].drop_duplicates(),
    on='restriction_id',
    how='left'
)

# Build features
features = df.groupby(['stop_id', 'route_id']).agg(
    restriction_count=('project_id', 'nunique'),
    road_diversity=('road', 'nunique'),
    closest_restriction=('distance_m', 'min')
).reset_index()

# Show project IDs
project_ids = df.groupby(
    ['stop_id', 'route_id']
)['project_id'].apply(
    lambda x: ', '.join(sorted(x.unique()))
).reset_index()

features = features.merge(
    project_ids,
    on=['stop_id', 'route_id'],
    how='left'
)

# Add stop names
features = features.merge(
    stops_df[['stop_id', 'stop_name']].drop_duplicates(),
    on='stop_id',
    how='left'
)

# Risk flag
features['high_risk'] = (
    features['closest_restriction'] <= 75
).astype(int)

# Reorder columns
features = features[
    [
        'stop_id',
        'stop_name',
        'route_id',
        'restriction_count',
        'project_id',
        'road_diversity',
        'closest_restriction',
        'high_risk'
    ]
]

# Sort results
features = features.sort_values(
    ['restriction_count', 'closest_restriction'],
    ascending=[False, True]
)

features.head(20)
```

```python
from pathlib import Path
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import LineString

# =========================================================
# CONFIG
# =========================================================
DATA_DIR = Path("data")
GTFS_DIR = DATA_DIR / "ttc_gtfs"
ROAD_DIR = DATA_DIR / "road_restrictions"

# =========================================================
# LOAD GTFS
# =========================================================
routes_df = pd.read_csv(GTFS_DIR / "routes.txt")
trips_df = pd.read_csv(GTFS_DIR / "trips.txt")
shapes_df = pd.read_csv(GTFS_DIR / "shapes.txt")

# =========================================================
# CLEAN SHAPES
# =========================================================
shapes_df["shape_pt_lat"] = pd.to_numeric(shapes_df["shape_pt_lat"], errors="coerce")
shapes_df["shape_pt_lon"] = pd.to_numeric(shapes_df["shape_pt_lon"], errors="coerce")
shapes_df["shape_pt_sequence"] = pd.to_numeric(shapes_df["shape_pt_sequence"], errors="coerce")

shapes_df = shapes_df.dropna(subset=[
    "shape_pt_lat", "shape_pt_lon", "shape_pt_sequence"
])

# =========================================================
# BUILD SHAPE LINES
# =========================================================
shape_list = []

for shape_id, group in shapes_df.groupby("shape_id"):
    group = group.sort_values("shape_pt_sequence")

    coords = list(zip(group["shape_pt_lon"], group["shape_pt_lat"]))

    if len(coords) >= 2:
        shape_list.append({
            "shape_id": shape_id,
            "geometry": LineString(coords)
        })

shape_gdf = gpd.GeoDataFrame(shape_list, geometry="geometry", crs="EPSG:4326")

# =========================================================
# ROUTE GEOMETRY
# =========================================================
route_shapes = trips_df[["route_id", "shape_id"]].drop_duplicates()
route_shapes = route_shapes.merge(shape_gdf, on="shape_id", how="inner")

route_gdf = gpd.GeoDataFrame(route_shapes, geometry="geometry", crs="EPSG:4326")

route_gdf = route_gdf.dissolve(by="route_id").reset_index()

# =========================================================
# LOAD ROAD RESTRICTIONS (KEEP skiprows=1)
# =========================================================
road_frames = []

for file in ROAD_DIR.glob("*.csv"):
    df = pd.read_csv(file, skiprows=1, low_memory=False)
    df.columns = df.columns.str.strip()
    road_frames.append(df)

road_df = pd.concat(road_frames, ignore_index=True)

print("Road raw shape:", road_df.shape)

# =========================================================
# CLEAN COORDINATES
# =========================================================
road_df["Latitude"] = pd.to_numeric(road_df["Latitude"], errors="coerce")
road_df["Longitude"] = pd.to_numeric(road_df["Longitude"], errors="coerce")

road_df = road_df.dropna(subset=["Latitude", "Longitude"])

# =========================================================
# FIX DATE FORMAT (YYYYMMDD)
# =========================================================
road_df["start_dt"] = pd.to_datetime(
    road_df["StartTime"].astype(str),
    format="%Y%m%d",
    errors="coerce"
)

road_df["end_dt"] = pd.to_datetime(
    road_df["EndTime"].astype(str),
    format="%Y%m%d",
    errors="coerce"
)

# =========================================================
# FILTER ACTIVE ROADS
# =========================================================
today = pd.Timestamp.now()
future = today + pd.Timedelta(days=7)

road_df = road_df[
    (road_df["start_dt"] <= future) &
    (road_df["end_dt"] >= today)
]

print("Road after time filter:", road_df.shape)

# =========================================================
# CREATE GEO DATAFRAME
# =========================================================
road_gdf = gpd.GeoDataFrame(
    road_df,
    geometry=gpd.points_from_xy(
        road_df["Longitude"],
        road_df["Latitude"]
    ),
    crs="EPSG:4326"
)

# =========================================================
# PROJECT TO METERS
# =========================================================
route_gdf = route_gdf.to_crs(epsg=32617)
road_gdf = road_gdf.to_crs(epsg=32617)

# =========================================================
# BUFFER ROUTES (IMPORTANT)
# =========================================================
route_gdf["geometry"] = route_gdf.geometry.buffer(300)

# =========================================================
# SPATIAL JOIN
# =========================================================
joined = gpd.sjoin(
    route_gdf,
    road_gdf,
    how="inner",
    predicate="intersects"
)

print("Routes:", len(route_gdf))
print("Road points:", len(road_gdf))
print("Matches:", len(joined))

# =========================================================
# FEATURE ENGINEERING
# =========================================================
if len(joined) > 0:

    route_features = joined.groupby("route_id").agg(
        shape_restriction_count=("ID", "nunique"),
        shape_road_diversity=("Road", "nunique"),
        restriction_density=("ID", "count")
    ).reset_index()

    route_features["shape_high_risk"] = (
        route_features["shape_restriction_count"] > 0
    ).astype(int)

else:
    route_features = pd.DataFrame(columns=[
        "route_id",
        "shape_restriction_count",
        "shape_road_diversity",
        "restriction_density",
        "shape_high_risk"
    ])

# =========================================================
# OUTPUT
# =========================================================
route_features = route_features.sort_values(
    "shape_restriction_count",
    ascending=False
)

print(route_features.head(20))
```
